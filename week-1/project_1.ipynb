```python
# For Simple Interest let
P = float(input("Enter Principal (P): "))
R = float(input("Enter Rate (R): "))
T = float(input("Enter Time (T): "))

A = P * (1 + (R / 100) * T)
print("Simple Interest:", A)
```

```python
# For Compound Interest let
P = float(input("Enter Principal (P): "))
R = float(input("Enter Rate (R): "))
n = float(input("Enter value for n: "))
t = float(input("Enter value for t: "))

A = P * (1 + (R / n)) ** (n * t)
print("Compound Interest: ", A)
```

![Flowchart](CompoundInterestFlowchart.png)

```python
# For Annuity Plan let
PMT = float(input("Enter Periodic Payment (PMT): "))
R = float(input("Enter Rate (R): "))
n = float(input("Enter value for n: "))
t = float(input("Enter value for t: "))

A = PMT * (((1 + (R / n)) ** (n * t)) - 1) / (R / n)
print("Annuity Plan: ", A)
```
