```python
import pandas as pd
import matplotlib.pyplot as plt  # for drawing the charts

df = pd.read_csv("AB_NYC_2019.csv") 

df = df.drop(columns=["id", "host_id", "latitude", "longitude", "last_review"]) # to remove columns we dont need
df = df.dropna() # to remove rows with missing values
df = df[df["price"] > 0] # to keep listings where the price is greater than 0

top_neighbourhoods = df["neighbourhood"].value_counts().head(10) # counts how many listings each neighbourhood has and keeps the top 10
top_neighbourhoods.plot(kind="bar") # creates a bar chart
plt.title("Top 10 Neighbourhoods")
plt.xlabel("Neighbourhood")
plt.ylabel("Number of Listings")
plt.show()

affordable = df[df["price"] < 500] # to keep listings where the price is less than 500
plt.hist(affordable["price"], bins=50) # creates a histogram with 50 divisions, each division is 10 as 500/50 is 10
plt.title("Price Distribution")
plt.xlabel("Price per Night ($)")
plt.ylabel("Number of Listings")
plt.show()

room_counts = df["room_type"].value_counts() # counts how many listings each room has
plt.pie(room_counts, labels=room_counts.index, autopct="%1.1f%%") # .index shows the names instead of the numbers and autopct shows the percent on each slice
plt.title("Room Type Breakdown")
plt.show()
```
