```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")
df = df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"]) # remove columns we dont need
df["Age"] = df["Age"].fillna(df["Age"].mean()) # fill missing age with average age
df["Embarked"] = df["Embarked"].fillna("Unknown") # fill missing embarked with Unknown
df = df[df["Fare"] > 0] # keep rows where fare is greater than 0
df = df[df["Fare"] < 300] # remove extreme outliers

top_ages = df["Age"].value_counts().head(10) # counts how many passengers each age has and keeps the top 10
top_ages.plot(kind="bar") # creates a bar chart
plt.title("Top 10 Most Common Ages")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()

affordable = df[df["Fare"] < 100] # keep fares less than 100
plt.hist(affordable["Fare"], bins=50) # creates a histogram with 50 divisions
plt.title("Fare Distribution")
plt.xlabel("Fare ($)")
plt.ylabel("Number of Passengers")
plt.show()

survived_counts = df["Survived"].value_counts() # counts how many survived vs died
plt.pie(survived_counts, labels=["Did not survive", "Survived"], autopct="%1.1f%%") # autopct shows percent on each slice
plt.title("Survival Breakdown")
plt.show()
```
